```python
from collections import UserDict
from copy import deepcopy


def logNDict_iter(my_dict):
    dic = deepcopy(my_dict)
    for sec_type in dic.keys():
        for sec_lambda in dic[sec_type].keys():
            for logN in dic[sec_type][sec_lambda].keys():
                yield sec_type, sec_lambda, logN, dic[sec_type][sec_lambda][logN]
                    
        
params = {
        'classic':{
            128:{
                1024:  29,
                2048:  56,
                4096:  111,
                8192:  220,
                16384: 440,
                32768: 880
            },
            192:{
                1024:  21,
                2048:  39,
                4096:  77,
                8192:  154,
                16384: 307,
                32768: 612
            },
            256:{
                1024:  16,
                2048:  31,
                4096:  60,
                8192:  120,
                16384: 239,
                32768: 478
            }
        },
        'quantum':{
           128:{
                1024:  25,
                2048:  51,
                4096:  101,
                8192:  202,
                16384: 411,
                32768: 827
            },
            192:{
                1024:  17,
                2048:  35,
                4096:  70,
                8192:  141,
                16384: 284,
                32768: 571
            },
            256:{
                1024:  13,
                2048:  27,
                4096:  54,
                8192:  109,
                16384: 220,
                32768: 443
            }
        },


    }
for i in logNDict_iter(params):
    print(i)
```

```python
def transform_params(params):
    for sec_type, sec_lambda, N, value in logNDict_iter(params):
        logN = int(np.log2(N))
        params[sec_type][sec_lambda][logN] = value
        params[sec_type][sec_lambda].pop(N)
        
    return params

params = transform_params(params)
params
```

We create a Objective Function that maximizes security. In order to equalize the probabilities of the objective function to take all the `logN`, we weight them according to rules that grant the following:

- Prefer a higher `quantum` security over `classic` security when possible.
- Prefer to use a higher level of lambda security `lambda`, when possible.
- Prefer to use a lower `logN` whenever possible. 

With this, we ensure the search of `logN` is done according to a better proportion to security

It is a knapsack problem, but you can also choose the backpack. There is not really a way to solve it and understand it is optimal. We want to ensure we can bring the necessary items, but the backpack is defined by the code that we have to optimize.


```python
MinLambda = 128
MaxLambda = 512


def generate_logN_objective_function():
    ObjectiveFunction = 0
    ##################################
    ## Measuring the impact of logN ##
    ##################################
    
    #######################################################
    ## B) Measuring the number of operations per vector  ##
    #######################################################
    
    dic_max_v = lambda x: np.array([v for _, v in x.items()]).max()
    adds_max = dic_max_v(ADDS)
    mults_max = dic_max_v(MULTS)
    rots_max = dic_max_v(ROTS)
    coeff_dic = lambda dic, coeff: {k: (v / coeff) for k, v in dic.items()} 
    ADDS_COEFF = coeff_dic(ADDS, adds_max)
    MULTS_COEFF = coeff_dic(MULTS, mults_max)
    ROTS_COEFF = coeff_dic(ROTS, rots_max)
    
    UNUSED_SPACE = {logN: (1 / (MAX_VECTOR_SIZE / (1 << (logN - 1)))) for logN in range(10, 16)}
    unused_max = np.array([v for k, v in UNUSED_SPACE.items()]).max()
    UNUSED_SPACE = {k: max(1, v)/unused_max for k, v in UNUSED_SPACE.items()}
    print(UNUSED_SPACE)
    l = []
    for sec_type, sec_lambda, logN, value in logNDict_iter(params):
        # Quantum security is better, and more restrictive, we prefer quantum if possible. 
        # We weight it by giving quantum a better rating
        sec_type_rate = params[sec_type][sec_lambda][logN] / params['classic'][sec_lambda][logN]
        # For the security parameter, we prefer higher levels of security as much as possible
        # Thus, we invert the potential weight of the parameters
        sec_lambda_rate = 1 - ((sec_lambda - MinLambda) / (MaxLambda - MinLambda))
        # A lower N guarantees better performance
        # A higher N guarantees better security
        logN_rate = (logN - MinLogN) / (MaxLogN - MinLogN)
        logN_rate = apply_logN_FL_score(logN_rate, logN_FL_score)
        op_rate = (ADDS_COEFF[logN] * 0.2) + (MULTS_COEFF[logN] * 0.5) + (ROTS_COEFF[logN] * 0.3)
        unused_space_rate = UNUSED_SPACE[logN]
        rate = sec_type_rate * sec_lambda_rate * logN_rate * op_rate * unused_space_rate
        l.append(rate)
    i = 0
    X = np.array(l)
    X = (X - X.min()) / (X.max() - X.min())
    

    
    PAD = 90
    print("SINCE WE MINIMIZE THE FUNCTION, LOWER IS BETTER")

    sec_lambda_prev = None 
    sec_type_prev = None
    for sec_type, sec_lambda, logN, value in logNDict_iter(params):
        if sec_lambda_prev != sec_lambda:
            print("-" * PAD)
            sec_lambda_prev = sec_lambda
        
        if sec_type_prev != sec_type:
            print("*" * PAD)
            print("Name             | TYPE | LMBD | LOGN | ADDS | MULS | ROTS | OPS_ | SPCE | TOTL | TOTS |")
            print("*" * PAD)
            sec_type_prev = sec_type
            
        # Quantum security is better, and more restrictive, we prefer quantum if possible. 
        # We weight it by giving quantum a better rating
        sec_type_rate = params[sec_type][sec_lambda][logN] / params['classic'][sec_lambda][logN]
        # For the security parameter, we prefer higher levels of security as much as possible
        # Thus, we invert the potential weight of the parameters
        sec_lambda_rate = 1 - ((sec_lambda - MinLambda) / (MaxLambda - MinLambda))
        # A lower N guarantees better performance
        # A higher N guarantees better security
        logN_rate = (logN - MinLogN) / (MaxLogN - MinLogN)
        logN_rate = apply_logN_FL_score(logN_rate, logN_FL_score)
        op_rate = (ADDS_COEFF[logN] * 0.2) + (MULTS_COEFF[logN] * 0.5) + (ROTS_COEFF[logN] * 0.3)
        unused_space_rate = UNUSED_SPACE[logN]
        rate = sec_type_rate * sec_lambda_rate * logN_rate * op_rate * unused_space_rate
        string = "|" + " %0.2f |" * 10
        print(NChoice[sec_type][sec_lambda][logN], 
              string % (sec_type_rate, 
                        sec_lambda_rate,
                        logN_rate,
                        ADDS_COEFF[logN],
                        MULTS_COEFF[logN],
                        ROTS_COEFF[logN],
                        op_rate,
                        unused_space_rate,
                        rate,
                        X[i]))
        ObjectiveFunction += NChoice[sec_type][sec_lambda][logN] * X[i]
        i += 1
        
    print("*" * PAD)
    
    ###################################
    ## Measuring the impact of logQi ##
    ###################################
    
    # The overall impact of each logQ needs to be equivalent to the one in logN
    mult_depth_rate = 1 / MAX_MULT_DEPTH
    # The overall impact of logQ needs to be equivalent to the one in logN
    q_rate = 1 / (MaxModuliSize - MinModuliSize)
    # The overall impact of slack needs to be balanced 
    slack_q_rate = (MinModuliSize) / (MaxModuliSize - MinModuliSize)
    for bq_choice, bqt_choice in zip(bQChoice, bQTChoice):
        #ObjectiveFunction += q_choice * q_rate * mult_depth_rate
        ObjectiveFunction += bq_choice * mult_depth_rate
        if LOGQ_RATE < 0.5:
            ObjectiveFunction -= bqt_choice * mult_depth_rate 
        else:
            ObjectiveFunction += bqt_choice * mult_depth_rate 
        #ObjectiveFunction -= slack_q_rate
        


    
    return ObjectiveFunction, l

logN_FL_score = LogNFL().simulate(security=SECURITY, performance=PERFORMANCE)

print("logN_FL_score: %f" % (logN_FL_score))

def apply_logN_FL_score(val, logN_FL_score):
    if 0.0 < val > 1.0:
        raise Exception("Expected value in the range 0.0 <= x <= 1.0, value is: %f" %(val))
    #print(logN_FL_score, val, (1 - val) * logN_FL_score, (1 - logN_FL_score) * val)
    return  (1 - val) * logN_FL_score + (1 - logN_FL_score) * val

print(apply_logN_FL_score(1.00, 0))
print(apply_logN_FL_score(0.29, logN_FL_score))
ObjectiveFunction, l = generate_logN_objective_function()
model += ObjectiveFunction
ObjectiveFunction
```

# Constraints

#### 1. Choose at most one $logN$
At most, we have to choose one of the `logN` with one security level $\lambda$ and one type of security `quantum` or `classic`.

```python
model += (lpSum(NChoice) == 1, "choose_only_one_logN")
```

#### 2. Maximum $logQ$ budget for each $logN$
Each `logN` has a `logQ` budget associated. We need to ensure that it is never surpassed for security according to homomorphic encryption standard.

```python
def get_maximum_logN_expr():
    expression = 0
    for sec_type in params.keys():
        for sec_lambda in params[sec_type].keys():
            for logN in params[sec_type][sec_lambda].keys():
                budget = params[sec_type][sec_lambda][logN]
                expression += NChoice[sec_type][sec_lambda][logN] * budget
    return lpSum(QChoice) <= expression 
logN_budget_expression = get_maximum_logN_expr()
model += (logN_budget_expression, 
          "sum_logQi_not_to_surpass_logN_budget")
```

#### Different $O_d$ per $logN$.
We may have a case where different multiplication depths $O_d$ are selected depending on the intermediate operations.

#### 3. Boolean Variables to check 
We create boolean variables that characterize if a $Q_i$ is used for a $LogN$. We will introduce as many $Q_i$ as the $max(O_{d}^{i}) \in logN$. We may not need to use them all depending on the logN that we use. For that, we introduce a rule that ensures that we only use as many as needed.

```python
def logQ_boolean_variables(max_mult_depth=MAX_MULT_DEPTH):

    expression = 0
    for sec_type in params.keys():
        for sec_lambda in params[sec_type].keys():
            for logN in params[sec_type][sec_lambda].keys():
                expression += NChoice[sec_type][sec_lambda][logN] * MULT_DEPTH_logN[logN]
    expression = expression == lpSum(bQChoice)
    return expression

expression = logQ_boolean_variables(MAX_MULT_DEPTH)
model += (expression,  "Set bQi variables to appropriate values depending on logN")
```

#### 4. Real Scale Parameter (`RQChoice`) to be in Real Parameter Range
We would like that `RQChoice` preserves the range `REAL_SCALE_MIN <= RQChoice <= REAL_SCALE_MAX`.

#### 5. The Parameters $Q_i \forall i \in \{0, N\}$ need to have at least `RQChoice` bits with $Q_0$

```python
for i in range(1, len(QChoice) -1):        
    model += (QChoice[i] + RQChoice <= QChoice[0],  "logQ0_has_n_bits_with_logQ%d"%( i))
```

#### 6. Ensure that: $logQ_0 = logQ_{O_{d}}$
We would like to have a good encryption budget and a good decryption budget, therefore $$B:logQ_0 = logQ_{O_{d}}$$

```python
model += (QChoice[0] == QChoice[len(QChoice) -1], "logQ0_is_equal_to_logQN")
```

#### 6. Ranges of $logQ_i$
Each logQi has to be in the range  $30 <= logQi <= 60$. The constraint is already imposed at variable declaration.

We use `bQChoice[i]` to determine, that a given Qi has to have value 0 if the given `logN` does not need from it.

We use `bQTChoice[i]` to determine thresholds. We penalize a variable once if used with `bQChoice[i]`, however, increasing its value withing the `[DEC_SCALE_MIN, DEC_SCALE_MAX_BOUND]` range does not penalize, thus, has freedom to use all the budget as best to match it with `logN`. Furthermore, bQTChoice

```python
for i in range(1, len(QChoice) -1):
    model += (QChoice[i] >= bQChoice[i] * DEC_SCALE_MIN,  "logQ%d_at_least_%d_bits" % (i, DEC_SCALE_MIN))
    model += (QChoice[i] <= bQChoice[i] * DEC_SCALE_MAX_BOUND,  "logQ%d_less_than_%d_bits" % (i, DEC_SCALE_MAX))
```

```python

for i in range(1, len(QChoice) - 1):
    model += (QChoice[i] >= bQTChoice[i] * DEC_SCALE_MIN_BOUND,  "logQ%d_at_least_%d_bits" % (i, DEC_SCALE_MIN_BOUND))
    
```
